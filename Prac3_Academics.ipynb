## Outlier detection using inter quantile range (IQR)

```python
q1=df.Age.quantile(0.25)
q3=df.Age.quantile(0.75)
q1,q3
```

```python
IQR=q3-q1
IQR
```

```python
lowe_limit=q1-1.5*IQR
upper_limit=q3+1.5*IQR
lowe_limit,upper_limit
```

```python
df[(df.Age<lowe_limit)|(df.Age>upper_limit)]
```

```python
df[(df.Age>lowe_limit)&(df.Age<upper_limit)]
```

## Detecting using Z score

```python
## z=(x-mean)/standarddeviation
```

```python
df['zscore']=(df.Age - df.Age.mean())/df.Age.std()
df
```

### These all are outliers.

```python
df[df['zscore']>3]
```

```python
df[df['zscore']<-3]
```
